```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
df = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Chris'],
    'income_2020': [50, 60, 60],
    'income_2021': [55, 65, 60],
    'income_2022': [60, 70, 60],
    'expenses_2020': [30, 40, 60],
    'expenses_2021': [35, 45, 60],
    'expenses_2022': [40, 50, 60],
    'intelligence': ['smart', 'dumb', 'really dumb']

})
new_df = pd.wide_to_long(df, 
    stubnames=['income', 'expenses'],
    i='id',
    j='year',
    sep='_',
    suffix='\\d+'
)
new_df = new_df.reset_index()
print('='*50)
b = pd.get_dummies(new_df[['name', 'intelligence', 'year']], drop_first=True, dtype=int)
b = pd.concat([b, new_df['expenses']], axis=1)


model = sm.OLS(new_df['income'], sm.add_constant(b))
results = model.fit()
print(results.summary())
```
