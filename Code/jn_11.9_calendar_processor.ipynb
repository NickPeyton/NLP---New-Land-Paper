```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import platform

if platform.node() == 'Nick_Laptop':
    drive = 'C'
elif platform.node() == 'MSI':
    drive = 'D'
else:
    drive = 'uhhhhhh'
    print('Uhhhhhhhhhhhhh')
os.chdir(f'{drive}:/PhD/DissolutionProgramming/LND---Land-Paper')

PROCESSED = 'Data/Processed'
RAW = 'Data/Raw'


calendar_df = pd.read_csv(f'{RAW}/calendar_of_particulars/individual_calendar_of_particulars.csv', encoding='utf-8')
```

```python
new_cal_df = pd.DataFrame(columns=['grant_num', 'grant_total', 'surname', 'forename', 'day', 'month', 'year',
       'in_devon'])

for i, row in calendar_df.iterrows():
    if pd.isna(row['surname']):
        continue
    grant_total = row['grant_total']
    surname = row['surname']
    forename = row['forename']

    if not pd.isna(row['surname_2']):
        grant_total = grant_total / 2
        surname_2 = row['surname_2']
        forename_2 = row['forename_2']
        new_row = pd.DataFrame({'grant_num': row['grant_num'], 'grant_total': grant_total,
                                        'surname': surname_2, 'forename': forename_2,
                                        'day': row['day'], 'month': row['month'], 'year': row['year'],
                                        'in_devon': row['in_devon']}, index=[0])
        new_cal_df = pd.concat([new_cal_df, new_row], ignore_index=True)
        print(f'Adding {surname_2} {forename_2} {grant_total} {row["day"]} {row["month"]} {row["year"]}')
    new_row = pd.DataFrame({'grant_num': row['grant_num'], 'grant_total': grant_total,
                                    'surname': surname, 'forename': forename,
                                    'day': row['day'], 'month': row['month'], 'year': row['year'],
                                    'in_devon': row['in_devon']}, index=[0])
    new_cal_df = pd.concat([new_cal_df, new_row], ignore_index=True)
```

```python
print(new_cal_df)
new_cal_df.to_csv(f'{PROCESSED}/calendar_recipients.csv', index=False, encoding='utf-8')
```
