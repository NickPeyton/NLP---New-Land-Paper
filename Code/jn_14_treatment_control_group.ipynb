```python
id_vars = ['unique_id', 'group_id', 'combined_id', 'metaphone_id', 'master_id']
calendar_df = pd.read_csv(f'{PROCESSED}/calendar_recipients_final.csv')
calendar_df['in_devon'] = calendar_df['in_devon'].str.replace('?', '0')
calendar_df['in_devon'] = calendar_df['in_devon'].fillna(0)
calendar_df['in_devon'] = calendar_df['in_devon'].astype(float)
calendar_df = calendar_df.loc[calendar_df['in_devon'] > 0]
treatment_id_dict = {}
grant_total_dict = {}
for id_var in id_vars:
    
    gdf = calendar_df.groupby(id_var).agg(
        {
            'grant_total': 'sum',
        })
    var_grant_total_dict = gdf['grant_total'].to_dict()
    grant_total_dict[id_var] = var_grant_total_dict

    treatment_id_dict[id_var] = [k for k, v in var_grant_total_dict.items()]
    print(len(treatment_id_dict[id_var]))
    
    
with open(f'{PROCESSED}/pretty_dict.json', 'r') as f:
    pretty_dict = json.load(f)
```

```python
wills_df = pd.read_csv(f'{PROCESSED}/ukda_pcc_wills_final.csv')
wills_df = wills_df.loc[wills_df['year'] < 1536]
control_id_dict = {}
significant_vars = []
for id_var in id_vars:
    id_type = id_var.split('_')[0]
    df = pd.read_csv(f'{SURNAMES}/{id_var}_subsidy_data_adjusted.csv')
    print(len(df))
    # Get share of wills for each surname id
    df['will_share_1536'] = df['id'].apply(
        lambda x: len(wills_df.loc[wills_df[id_var] == x]) / len(wills_df))
    treatment_list = treatment_id_dict[id_var]
    df['treatment'] = df['id'].apply(
        lambda x: 1 if x in treatment_list else 0)

    # Regression to find relevant variables for treatment assignment
    rdf = df.copy()
    x_vars = ['will_share_1536', 'avg_val_1524', 'ln_count_1524', 'ln_max_val_1524', 'ln_parish_monastic_land', 'parish_distmkt', 'parish_distriver', 'parish_pop_density_1524', 'ln_parish_avg_value_1524']
    for var in x_vars:
        x_max = rdf[var].max()
        x_min = rdf[var].min()
        # Normalize variables
        rdf[var] = (rdf[var] - x_min) / (x_max - x_min)

    x = rdf[x_vars]
    x[['ln_parish_monastic_land', 'parish_distmkt', 'parish_distriver']].fillna(0, inplace=True)
    x.dropna(inplace=True)
    x.rename(columns=pretty_dict, inplace=True)
    x = sm.add_constant(x)
    y = rdf['treatment']
    y = y.loc[x.index]
    model = sm.Logit(y, x)
    result = model.fit(cov_type='HC3', maxiter=1000, disp=False)
    # Grab variables with p-value < 0.10
    significant_vars += result.pvalues[result.pvalues < 0.1].index.tolist()
    significant_vars.remove('const')  # Remove constant term    
    coef_df = pd.DataFrame(result.summary().tables[1].data)
    coef_df.columns = coef_df.iloc[0]
    coef_df = coef_df[2:]
    coef_df[['coef', '[0.025', '0.975]']] = coef_df[['coef', '[0.025', '0.975]']].apply(pd.to_numeric, errors='coerce')

    
    # # Set index to variable names 
    coef_df = coef_df.set_index(coef_df.columns[0])
    variables = list(coef_df.index.values)
    # print(coef_df)

    # Change datatype from object to float
    # coef_df = coef_df.astype(float)
    # Add variables column to dataframe
    coef_df['variables'] = variables
    
    # Set sns plot style back to 'poster'
    # This will make bars wide on plot
    sns.set_context("poster")

    # Define figure, axes, and plot
    fig, ax = plt.subplots(figsize=(15, 10))
    errors = coef_df['coef'] - coef_df['[0.025']
    
    # Append errors column to dataframe
    coef_df['errors'] = errors

    # Error bars for 95% confidence interval
    # Can increase capsize to add whiskers
    coef_df.plot(x='variables', y='coef', kind='bar',
                 ax=ax, color='none', fontsize=22, 
                 ecolor='steelblue',capsize=0,
                 yerr='errors', legend=False)
    
    # Set title & labels
    plt.title(f'{id_type.title()} ID Treatment Predictors',fontsize=30)
    ax.set_ylabel('Coefficients',fontsize=22)
    ax.set_xlabel('',fontsize=22)
    
    # Coefficients
    ax.scatter(x=np.arange(coef_df.shape[0]), 
            marker='o', s=80, 
            y=coef_df['coef'], color='steelblue')

    # Label x-axis with variable names
    ax.set_xticks(np.arange(len(coef_df)))
    ax.set_xticklabels(coef_df['variables'], rotation=45, ha='right', fontsize=18)
    # Line to define zero on the y-axis
    ax.axhline(y=0, linestyle='--', color='red', linewidth=1)
    # Save figure
    plt.tight_layout()
    plt.savefig(f'{IMAGES}/treatment_predictors_{id_var}_adjusted.png', dpi=300)
```

```python
wills_df = pd.read_csv(f'{PROCESSED}/ukda_pcc_wills_final.csv')
wills_df = wills_df.loc[wills_df['year'] < 1536]
control_id_dict = {}
significant_vars = []
for id_var in id_vars:
    df = pd.read_csv(f'{SURNAMES}/{id_var}_subsidy_data_adjusted.csv')
    print(len(df))
    # Get share of wills for each surname id
    df['will_share_1536'] = df['id'].apply(
        lambda x: len(wills_df.loc[wills_df[id_var] == x]) / len(wills_df))
    treatment_list = treatment_id_dict[id_var]
    df['treatment'] = df['id'].apply(
        lambda x: 1 if x in treatment_list else 0)

    # Matching each treatment observation with a control observation, without replacement
    predictors = [
        'ln_tot_val_1524',
        'ln_count_1524',
        'will_share_1536', 
        'ln_max_val_1524', 
        'ln_parish_avg_value_1524',
        'ln_parish_monastic_land']

    treatment_df = df.loc[df['treatment'] == 1].copy()
    treatment_df.fillna(0, inplace=True)
    control_df = df.loc[df['treatment'] == 0].copy()
    control_df.fillna(0, inplace=True)
    # Normalize all predictors to take values between 0 and 1 based on max and min of whole dataset
    for predictor in predictors:
        treatment_df[predictor] = treatment_df[predictor] - df[predictor].min()
        treatment_df[predictor] = treatment_df[predictor] / (df[predictor].max() - df[predictor].min())
        control_df[predictor] = control_df[predictor] - df[predictor].min()
        control_df[predictor] = control_df[predictor] / (df[predictor].max() - df[predictor].min())

    X_treatment = treatment_df[predictors].values
    X_control = control_df[predictors].values

    dist_matrix = pairwise_distances(X_treatment, X_control, metric='euclidean')

    available_control_indices = list(range(len(X_control)))
    matched_control_indices = []
    for i in range(len(X_treatment)):
        if len(available_control_indices) == 0:
            print('Oar Noar')
            break
        nearest_index = np.argmin(dist_matrix[i, available_control_indices])
        matched_control_indices.append(available_control_indices[nearest_index])
        del available_control_indices[nearest_index]

    matched_control_df = control_df.iloc[matched_control_indices].copy()


    treatment_df = treatment_df[predictors + ['id', 'treatment']].copy()
    treatment_df.rename(columns={x: f'treatment_{x}' for x in treatment_df.columns}, inplace=True)
    treatment_df.reset_index(drop=True, inplace=True)
    matched_control_df = matched_control_df[predictors + ['id', 'treatment']].copy()
    matched_control_df.rename(columns={x: f'control_{x}' for x in matched_control_df.columns}, inplace=True)
    matched_control_df.reset_index(drop=True, inplace=True)
    # Ensure that the treatment and matched control dataframes have the same number of rows   
    if len(treatment_df) != len(matched_control_df):
        print(f'Length mismatch: {len(treatment_df)} treatment vs {len(matched_control_df)} control')
        continue

    control_id_list = matched_control_df['control_id'].tolist()
    control_id_dict[id_var] = control_id_list
    # # Concatenate treatment and matched control dataframes along axis 1, preserving predictor data from both
    # combined_df = pd.concat([treatment_df, matched_control_df], axis=1)
    # print(combined_df.head())
    

matched_treatment_control_dict = {}
for id_var in id_vars:
    df = pd.read_csv(f'{SURNAMES}/{id_var}_subsidy_data_adjusted.csv')
    # Get share of wills for each surname id
    df['will_share_1536'] = df['id'].apply(
        lambda x: len(wills_df.loc[wills_df[id_var] == x]) / len(wills_df))
    treatment_list = treatment_id_dict[id_var]
    df['treatment'] = df['id'].apply(
        lambda x: 1 if x in treatment_list else 0)
    treatment_df = df.loc[df['treatment'] == 1].copy()
    treatment_df.fillna(0, inplace=True)
    control_df = df.loc[df['treatment'] == 0].copy()
    control_df.fillna(0, inplace=True)
    # Normalize all predictors to take values between 0 and 1 based on max and min of whole dataset
    for predictor in predictors:
        treatment_df[predictor] = treatment_df[predictor] - df[predictor].min()
        treatment_df[predictor] = treatment_df[predictor] / (df[predictor].max() - df[predictor].min())
        control_df[predictor] = control_df[predictor] - df[predictor].min()
        control_df[predictor] = control_df[predictor] / (df[predictor].max() - df[predictor].min())

    X_treatment = treatment_df[predictors].values
    X_control = control_df[predictors].values

    dist_matrix = pairwise_distances(X_treatment, X_control, metric='euclidean')

    available_control_indices_2 = list(range(len(X_control)))
    matched_groups = {}
    for i in range(len(X_treatment)):
        if len(available_control_indices_2) == 0:
            print('Oar Noar')
            break
        matched_groups[int(treatment_df['id'].iloc[i])] = i
        for j in range(4):
            nearest_index = np.argmin(dist_matrix[i, available_control_indices_2])
            control_id = int(control_df['id'].iloc[available_control_indices_2[nearest_index]])
            matched_groups[control_id] = i
            del available_control_indices_2[nearest_index]
    matched_treatment_control_dict[id_var] = matched_groups

# Save dicts of treatment and control ids to .json files
with open(f'{SURNAMES}/treatment_id_dict_adjusted.json', 'w') as f:
    json.dump(treatment_id_dict, f)
with open(f'{SURNAMES}/control_id_dict_adjusted.json', 'w') as f:
    json.dump(control_id_dict, f)
with open(f'{SURNAMES}/matched_groups_adjusted.json', 'w') as f:
    json.dump(matched_treatment_control_dict, f)
```

#### Applying our treatment and control groups to the surname data

```python
for id_var in id_vars:
    id_type = id_var.split('_')[0]
    file_path = f'{SURNAMES}/{id_var}_subsidy_data_adjusted.csv'
    df = pd.read_csv(file_path)
    df['id'] = df['id'].astype(int)  # Ensure id is int for matching
    treatment_list = treatment_id_dict[id_var]

    control_list = control_id_dict[id_var]
    group_dict = matched_treatment_control_dict[id_var]
    df[f'treatment'] = df['id'].apply(
        lambda x: 1 if x in treatment_list else 0)
    df[f'control'] = df['id'].apply(
        lambda x: 1 if x in control_list else 0)
    df['control2_group'] = None
    df['control2_group'] = df['id'].apply(lambda x: group_dict.get(int(x), None))
    df.to_csv(file_path, index=False)
```
