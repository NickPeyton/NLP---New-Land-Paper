#### Regressions

```python
id_vars = ['unique_id', 'group_id', 'combined_id', 'metaphone_id', 'master_id']

with open(f'{PROCESSED}/pretty_dict.json', 'r') as f:
    pretty_dict = json.load(f)

for id_var in id_vars:
    wdf = pd.read_csv(f'{SURNAMES}/will_panel_{id_var}.csv')
    # wdf = wdf.drop(columns=['lag_nan', 'lag_1.0'])
    xvars = [col for col in wdf.columns if 'lead_' in col or 'lag_' in col]
    xvars.remove('lead_1.0')
    xvars.remove('lag_nan')
    wdf['id'] = wdf['id'].astype(str)
    wdf['period'] = wdf['period'].astype(str)
    grouped = wdf.groupby('id')
    wdf_demeaned = wdf[xvars + ['will_share']] - grouped[xvars + ['will_share']].transform('mean')
    x = wdf_demeaned[xvars]
    x = sm.add_constant(x)  # Add a constant term for the intercept
    y = wdf_demeaned['will_share']
    results = sm.OLS(y, x).fit(cov_type='HC3')
    print(results.summary())
```
